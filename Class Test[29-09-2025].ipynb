```python
#Check if a day is weekday or weekend
weekday = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
day = input("Enter a day: ")
if day in weekday:
    print(f"{day} is a weekday.")
else:
    print(f"{day} is a weekend.")
```
