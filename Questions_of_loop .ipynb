```python
#Find the largest number in the list [4, 9, 2, 10, 7] using a for loop.
list = [4, 9, 2, 10, 7]
largest = list[0]
for num in list:
    if num > largest:
        largest = num
print("Largest number:",largest)
```

```python
#Count how many even and odd numbers are present in [1, 2, 3, 4, 5, 6].
numbers = [1, 2, 3, 4, 5, 6]
even_count = 0
odd_count = 0
for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
print(f"Even numbers: {even_count}")
print(f"Odd numbers: {odd_count}")
```

```python
#Use a for loop to create a new list containing squares of numbers from 1 to 5.
for i in range(1,6):
    print(f'{i} = {i*i}')
```

```python
#Check if a day is weekday or weekend.
day = str(input("Enter a day"))
weekends = ["Saturday", "Sunday","saturday","sunday"]
if day in weekends:
    print(f"{day} is a weekend.")
else:
    print(f"{day} is a weekday.")
print()
```

```python
#Write a program that asks the user to input a number and keeps asking for input until they enter a number greater than 10.
num = int(input("Enter a number greater than 10"))
while num <= 10:
    print("Try again!")
    num = int(input("Enter a number greater than 10"))
print(f'Good job! your entered {num}.')
```

```python
#Write a program to find the sum of all positive integers less than or equal to a given number n.
n = int(input("Enter numbers:"))
total = 0
i = 1
while i <= n:
    total += i
    i += 1
print(f'The sum of numbers from 1 to {n} is {total}.')
```

```python
#Write a program that calculates the factorial of a number entered by the user using a while loop.
n = int(input("Enter a number to calculate its factorial: "))
factorial = 1
i = 1
while i <= n:
    factorial *= i
    i += 1
print(f"The factorial of {n} is {factorial}.")
```
